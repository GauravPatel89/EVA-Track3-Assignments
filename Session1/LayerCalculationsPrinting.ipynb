```python
def showLayerCalulations(inSize,kernelSize,targetSize):
    # first findout by how much the i/p size decrease for given kerneSize
    print('Layer No.   |    i/p    |   Operation   |    o/p')
    sizeReductionFactor = kernelSize - 1
    layerNo = 1
    while inSize > targetSize:
        layerOutSize = inSize - sizeReductionFactor
        print('{:^12}|  {:^3}x{:^3}  |   {}x{} Conv    |  {:^3}x{:^3}'.format(layerNo,inSize,inSize,kernelSize,
                                                                             kernelSize,layerOutSize,layerOutSize))
        inSize = layerOutSize
        layerNo+=1
        
        
showLayerCalulations(199,3,1)
        
                                                                        
    
```
